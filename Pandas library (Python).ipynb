### loc

```python
printf("Labels:\n\tindex_labels={}, column_labels={}".format(df.index.tolist(), df.columns.tolist()))

#labels means indices or column_names
#non_label means list of labels or slice of labels

# loc syntax
#     * dataFrame.loc[I,J]; I,J in {labels, list_of_labels, slice_of_labels, boolean_series_or_list}
#     * series.loc[I] 

### ONE RULE ###
#*******************************************#
#  OUTPUT_DIMENSION = 2 - NUMBER_OF_LABELS  #
#*******************************************#

#        ___2 LABELS___ -> output is (2-2=0) 0-dim, i.e. scalar
#        ___1 LABELS___ -> output is (2-1=0) 1-dim, i.e. Series
#        ___0 LABELS___ -> output is (2-0=2) 2-dim, i.e. DataFrame



# ___2 LABELS___

# label,label
printf(df.loc["a","B"]) # scalar


# ___1 label___

#label, non_label (list of labels)
df.loc["a", ["B"]] # Series
#label, non_label (slice of labels)
df.loc["a", "A":"B"] # Series
# non_label (list of labels), label
df.loc[["a"], "B"] # Series
#non_label (slice of labels), label
df.loc["a":"b", "C"] # Series


# ___0 labels___

# non_label (list of labels), non_label (list of labels)
df.loc[["a","b"], ["A","C"]] # DataFrame
# non_label (list of labels), non_label (slice of labels)
df.loc[["a","b"], "A":"C"] # DataFrame
# non_label (slice of labels), non_label (list of labels)
df.loc["a":"b", ["A","C"]] # DataFrame
# non_label (slice of labels), non_label (slice of labels)
df.loc["a":"b", "A":"C"] # DataFrame


# for loc non specified dimensions are equal to ':'
df.loc["a"] == df.loc["a",:] #output=Series
df.loc[["b"]] == df.loc[["b"], :] # ouput=DataFrame

# slice start:end IINCLUDES the end, unlike python slices!
df.loc["a":"b"] #includes index"b" too
df.loc[:, "B":"C"] #includes columns "C" too



# Another way to determine output dimension of loc

# labels are 0 dimension
# list, slice and series are 1 dimension
# output_dimension of loc[I,J] is dimension_of_I + dimension_of_J

# output_dimension = 0 => scalar
# output_dimension = 1 => Series
# output_dimension = 2 => DataFrame
```
