## Merging DataFrames

```python
joy_df=pd.DataFrame({
                    "name":["Alice", "Bob", "Charlie"],
                     "hobby":["reading","gaming","painting"],
                        "age":[25,12,15]
                    }).set_index("name")
```

```python
joy_df
```

```python
work_df=pd.DataFrame({
                    "name":["Bob", "Charlie", "David"],
                     "work":["artist","student","writer"],
                        "age":[12,15,30]
                    }).set_index("name")
```

```python
work_df
```

```python
# outer-join (union, A u B) -> union of both the indices
outer_join=pd.merge( work_df, joy_df, how="outer", left_index=True, right_index=True)
```

```python
outer_join
```

```python
# inner-join (intersection, A n B) -> intersection of both the indices
inner_join=pd.merge(work_df, joy_df, how="inner", left_index=True, right_index=True)
```

```python
inner_join
```

```python
## left-join (A) -> only left index
working_people=left_join=pd.merge(work_df, joy_df, how="left", left_index=True, right_index=True)
```

```python
working_people
```

```python
## right-join (B) -> only right index
fun_people=right_join=pd.merge(work_df, joy_df, how="right", left_index=True, right_index=True)
```

```python
fun_people
```

```python
#merge on non-indices
joy_df.reset_index(inplace=True) #okay to use, as index has a name
work_df.reset_index(inplace=True) #okay to use, as index has a name
age_join=pd.merge(work_df, joy_df, how="outer", left_on="age", right_on="age")
```

```python
def all_gold(row):
    data=row.loc[["01 !", "01 !.1"]]
    return pd.Series({"Total Gold":np.sum(data)}) #return a new series #i.e. one new column "Total Gold"
```

```python
# The axis agument for apply method can be thought of as whether to apply change (by the function argument)
# along row (axis=0) or column (axis=1) of the input dataframe
```

```python
df.apply(all_gold, axis=1).head() #axis=1 across rows, axis=0 across columns!
# df.apply(func(row/column), axis=1/0)
```

```python
def all_gold2(row):
    data=row.loc[["01 !", "01 !.1"]]
    row.loc["Total Gold"]=np.sum(data)
    return row #return modified input row, effectively returning the input df with new column "Total Gold"
```

```python
df=df.apply(all_gold2, axis=1)
df.head()
```

```python
def all_sum(col):
    data=col
    col.loc["Total"]=np.sum(data)
    return col #return modified input col, effectively returning the input df with new row "Total"
```

```python
df2=df.iloc[0:len(df)-1].apply(all_sum, axis=0) #apply along a column
```

```python
df2.tail(5) # new "Total" row is added
```
